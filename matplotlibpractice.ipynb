```
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import io



```

```
df=pd.read_csv(io.BytesIO(uploaded['company_sales_data.csv']))
print(df.head())
```

```
profitList=df['total_profit'].tolist()
monthList=df['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.xticks(monthList)
plt.ylabel('Profit in dollar')
plt.title('Company profit per month')
plt.yticks([100000,200000,300000,400000,500000])
plt.show()
```

```
unitList=df['total_units'].tolist()
plt.plot(monthList,unitList,label='Profit data of last year',color='maroon',
         marker='o',markerfacecolor='k',linestyle='--',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')
plt.xticks(monthList)
plt.title('Units Sold')
plt.show()
```

```
facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()
toothpaste=df['toothpaste'].tolist()
bathingsoap=df['bathingsoap'].tolist()
shampoo=df['shampoo'].tolist()
moisturizer=df['moisturizer'].tolist()

plt.plot(monthList,facecream,label='facecream sales',marker='o',markerfacecolor='k')
plt.plot(monthList,facewash,label='facewash sales',marker='o',markerfacecolor='k')
plt.plot(monthList,toothpaste,label='toothpaste sales',marker='o',markerfacecolor='k')
plt.plot(monthList,bathingsoap,label='bathingsoap',marker='o',markerfacecolor='k')
plt.plot(monthList,shampoo,label='shampoo',marker='o',markerfacecolor='k')
plt.plot(monthList,moisturizer,label='moisturizer',marker='o',markerfacecolor='k')

plt.xlabel('month in number')
plt.ylabel('units sold')
plt.title('units sold by product')
plt.xticks(monthList)
plt.legend(loc='upper left')

plt.show()
```

```
plt.scatter(monthList,toothpaste,label='toothpaste sales data')
plt.xlabel('month in number')
plt.xticks(monthList)
plt.ylabel('units sold')
plt.title('toothpaste sales')
plt.legend(loc='upper left')
plt.grid(True,linestyle="--")
plt.show()
```

```
plt.bar(monthList,bathingsoap)
plt.xlabel('month in number')
plt.xticks(monthList)
plt.ylabel('units sold')
plt.grid(True,linestyle='--')
plt.title('Bathing soap sales')
plt.show()
```
