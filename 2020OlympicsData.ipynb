```python
import spacy
nlp = spacy.load("en_core_web_sm")

import io
ff = io.open('/kaggle/input/2020summerolympics/Mod2_2020Olympics.txt', 'r', errors='ignore')
doc = nlp(ff.read())
ff.close()
```

```python
from spacy import displacy

for ent in doc.ents:
    displacy.render(ent, style="ent")
```

```python
for ent in doc.ents:
    if ent.label_ == 'PERSON':
        displacy.render(ent, style='ent')
```

```python
Oly2020_names = []
for ent in doc.ents:
    if ent.label_ == 'PERSON':
        Oly2020_names.append(str(ent))
        
sorted_list = sorted(Oly2020_names, key=str)

from collections import Counter
import numpy as np

new_list = np.asarray(Counter(sorted_list).most_common(20))

print(new_list)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.barh(new_list[:,0], [int(x) for x in new_list[:,1]], color='blue', align='center')
ax.invert_yaxis()
ax.set_xlabel('Count of Name Occurrences in 2020 Olympics Wikipedia Article')
ax.grid()

plt.show()
```
